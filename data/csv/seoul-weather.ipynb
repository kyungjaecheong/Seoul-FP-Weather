```python
import numpy as np
import pandas as pd
```

```python
# 원본 data 불러오기
df0 = pd.read_csv('서울시_기상개황_원본.csv')
df0
```

```python
#행렬 변환
df1 = df0.T.reset_index(drop=False)
df1.head()
```

```python
#column 정의
columns = ['year_month','avgTa','maxTa','minTa','sumRn','avgRhm','avgPs','sumSsHr','avgWs']
df2 = df1.copy()
df2.columns = columns
df2.head()
```

```python
# 첫 두 행 제거
df3 = df2.iloc[2:,:].reset_index(drop=True)
df3.head()
```

```python
#year 행, month 행 생성
df4 = df3.copy()
df4['year'] = df4.year_month.str.split('. ').str[0].astype(int)
df4['month'] = df4.year_month.str.split('. ').str[1].astype(int)
df4.head()
```

```python
#year_month 표기 형식 변환
df5 = df4.copy()
df5.year_month = df5.year_month.str.replace('. ','-',regex=False)
df5.head()
```

```python
#column 순서 변경
df6 = df5.copy()
df6 = df6[['year_month','year','month','avgTa','maxTa','minTa','sumRn','avgWs','avgRhm','sumSsHr','avgPs']]
df6.head()
```

```python
#중간 점검용 data 저장
df6.to_csv('weather-temp-data.csv', index=False)
```

```python
#data test
df_test = pd.read_csv('weather-temp-data.csv')
df_test.info()
```

```python
#숫자형 아닌 data 추출
df_not_num = df_test[~df_test.sumRn.str.contains(r'\d+')|~df_test.sumSsHr.str.contains(r'\d+')]
df_not_num
```

```python
#결측치 채우기(기준: 서울(108) 30년(1991~2020) 월별평년값)
#1월 강수량 : 16.8 mm
#10월 강수량 : 52.2 mm
#1월 일조시간 : 169.6 hr
df_not_num_fix = df_not_num.copy()
df_not_num_fix.loc[204,'sumRn'] = 16.8
df_not_num_fix.loc[225,'sumRn'] = 52.2
df_not_num_fix.loc[228,'sumSsHr'] = 169.6
df_not_num_fix
```

```python
#덮어쓰기
df_test_2 = df_test.copy()
df_test_2.iloc[df_not_num.index] = df_not_num_fix
df_test_2.iloc[df_not_num.index]
```

```python
# 최종 data 저장
df_test_2.to_csv('seoul-weather-month.csv', index=False)
```

대시보드 데이터 추가 생성을 위한 수정

```python
#평년값 데이터 불러오기
ayv = pd.read_csv('서울시_월별평년값_1991_2020.csv')
ayv
```

```python
# 필요없는 column 제거
ayv1 = ayv.iloc[:,2:]
ayv1
```

```python
# 월별평년값 데이터 저장
ayv1.to_csv('seoul-average-month.csv',index=False)
```

```python
# 계절별 평균 데이터 생성을 위한 group by
ayv2 = ayv1.groupby(by='season').mean().round(1)
ayv2
```

```python
#month column 제거
ayv3 = ayv2.iloc[:,1:].reset_index(drop=False)
ayv3
```

```python
#보기좋게 행 정렬
ayv4 = ayv3.reindex([1,2,0,3]).reset_index(drop=True)
ayv4
```

```python
# 계절 평년데이터 저장
ayv4.to_csv('seoul-average-season.csv', index=False)
```
